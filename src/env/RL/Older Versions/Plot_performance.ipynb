```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from matplotlib import style
style.use("ggplot")

df = pd.read_csv('Leader_perspective.csv').round(2)
df1 = df
fig = plt.figure(figsize =(20, 10))
ax = fig.add_axes([0.1, 0.1, 0.6, 0.75])
corr = 0.4
Inspection = {t:(t,df["Inspection"][t]) for t in range(df.shape[0]) if df["Inspection"][t] == 1}
Maintenance = {t:(t,df["Maintenance"][t]) for t in range(df.shape[0]) if df["Maintenance"][t] == 1}
ax.plot(range(df.shape[0]), df["Performance"], 'k--', linewidth = 1.5, label = "Equilibrium solution for both parties")
ax.plot([Inspection[t][0] for t in Inspection], [Inspection[t][1] for t in Inspection], 'rs', label = "Inspection actions")
ax.plot([Maintenance[t][0] for t in Maintenance], [Maintenance[t][1] for t in Maintenance], 'b^', label = "Maintenance actions")
ax.set_xlabel("Period", size=15)
ax.set_ylabel("Road's Performance", size=15)
ax.legend(bbox_to_anchor=(1.05, 1), loc='best', borderaxespad=0.)
plt.suptitle("Leader's perspective" , fontsize=15)
plt.grid(True)
plt.savefig('Leader perspective.png')
# plt.show()
# plt.close(fig)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from matplotlib import style
style.use("ggplot")

df = pd.read_csv('Best_sol.csv').round(2)
df3 = df
fig = plt.figure(figsize =(20, 10))
ax = fig.add_axes([0.1, 0.1, 0.6, 0.75])
corr = 0.4
Inspection = {t:(t,df["Inspection"][t]) for t in range(df.shape[0]) if df["Inspection"][t] == 1}
Maintenance = {t:(t,df["Maintenance"][t]) for t in range(df.shape[0]) if df["Maintenance"][t] == 1}
ax.plot(range(df.shape[0]), df["Performance"], 'k--', linewidth = 1.5, label = "Equilibrium solution for both parties")
ax.plot([Inspection[t][0] for t in Inspection], [Inspection[t][1] for t in Inspection], 'rs', label = "Inspection actions")
ax.plot([Maintenance[t][0] for t in Maintenance], [Maintenance[t][1] for t in Maintenance], 'b^', label = "Maintenance actions")
ax.set_xlabel("Period", size=15)
ax.set_ylabel("Road's Performance", size=15)
ax.legend(bbox_to_anchor=(1.05, 1), loc='best', borderaxespad=0.)
plt.suptitle("Performance, inspection and maintenance actions for best solution found" , fontsize=15)
plt.grid(True)
plt.show()
# plt.savefig('Tradeoff.png')
# plt.close(fig)
```

```python
fig, axs = plt.subplots(3, sharex=True, sharey=True, figsize =(15, 15))
df = [df1,df2,df3]

Title_ = ["Leader's perspective","Follower's perspective","Tradeoff solution"]
for i in range(3):
    Inspection = {t:(t,df[i]["Inspection"][t]) for t in range(df[i].shape[0]) if df[i]["Inspection"][t] == 1}
    inspect = [df[i]["Inspection"][t] for t in range(df[i].shape[0]) if df[i]["Inspection"][t] == 1]
    maint = [df[i]["Maintenance"][t] for t in range(df[i].shape[0]) if df[i]["Maintenance"][t] == 1]
    print(f'Solution {i} has {sum(inspect)} inspections and {int(sum(maint))} maintenances')
    Maintenance = {t:(t,df[i]["Maintenance"][t]) for t in range(df[i].shape[0]) if df[i]["Maintenance"][t] == 1}
    axs[i].plot([t  for t in range(df[i].shape[0])], df[i]["Performance"], 'k--', linewidth = 1.5, label = "Equilibrium solution for both parties (road's performance)")
    axs[i].plot([Inspection[t][0] for t in Inspection], [Inspection[t][1] for t in Inspection], 'rs', label = "Inspection actions")
    axs[i].plot([Maintenance[t][0] for t in Maintenance], [Maintenance[t][1] for t in Maintenance], 'b^', label = "Maintenance actions")
    axs[i].set_title(Title_[i])
    if i == 0:        
        fig.legend(bbox_to_anchor=(0.525, 0.95), loc='center', borderaxespad=0)
    for ax in axs.flat:
        ax.set(xlabel='Period', ylabel='Performance')
    # Hide x labels and tick labels for top plots and y ticks for right plots.
    for ax in axs.flat:
        ax.label_outer()
    plt.xticks(range(df[i].shape[0]))
plt.savefig('Thy_plots.png')
```
